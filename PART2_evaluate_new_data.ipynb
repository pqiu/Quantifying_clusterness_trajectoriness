```python
def load_your_data(dir, pca_dim = 20):
    df = pd.read_csv(dir, index_col=0)
    npy_array = df.values
    m,n = npy_array.shape
    if n > 20:
        pca = PCA(n_components=pca_dim,random_state=42)
        embedding = pca.fit_transform(npy_array)
        return embedding
    else:
        return npy_array
    
```
